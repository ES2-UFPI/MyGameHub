# Imports

```python
import pandas as pd
import pickle
import torch
import os
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertTokenizer, BertModel
```

# Calculando Embeddings

```python
df = pd.read_csv('../data/processed_data.csv')

embeddings_file = '../embeddings/embeddings.pkl'

def generate_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()
    return embeddings.numpy()

if os.path.exists(embeddings_file):
    with open(embeddings_file, 'rb') as f:
        embeddings_list = pickle.load(f)
else:
    game_descriptions = df['game_description'].tolist()
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')

    embeddings_list = []
    for description in game_descriptions:
        try:
            embeddings = generate_embeddings(description)
            embeddings_list.append(embeddings)
        except Exception as e:
            print(f"Erro ao calcular embeddings para a descrição: {description}. Detalhes do erro: {e}")
    
    with open(embeddings_file, 'wb') as f:
        pickle.dump(embeddings_list, f)
```
